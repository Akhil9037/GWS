```python
#For h pluss and h cross polarization of the gravitational wave using.  arXiv no.......
# I forgot the paper number
# for the eccenticity e= 0.1


import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = 10  # frequency (Hz)
e = 0.1  # eccentricity
T = 10  # total time (s)
dt = 1e-3  # time step (s)
t = np.arange(0, T, dt)  # time array

# h+ Polarization
def h_plus_polarization(t, f, e):
    return (-1 / 2) * (1 + np.cos(e * np.pi / 2)) * np.cos(2 * np.pi * f * t)

h_plus = h_plus_polarization(t, f, e)

# hx Polarization
def h_cross_polarization(t, f, e):
    return (-1 / 2) * (1 + np.cos(e * np.pi / 2)) * np.sin(2 * np.pi * f * t)

h_cross = h_cross_polarization(t, f, e)

# Plot Strain vs Time
fig, axs = plt.subplots(2)
fig.suptitle('Gravitational Wave Signal - h+ and hx Polarizations')
axs[0].plot(t, h_plus, label='h+')
axs[0].set_ylabel('h+ Polarization')
axs[1].plot(t, h_cross, label='hx')
axs[1].set_ylabel('hx Polarization')
axs[1].set_xlabel('Time (s)')
plt.show()
```
